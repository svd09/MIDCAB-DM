```stata
// want new variables for age and bmi that are more meaningful

gen age10 = age/10

gen bmi5 = bmi/5


```

```stata
// understand lvfunction

tab lvfunction

tab1 sex diabetes

gen female = .
replace female = 1 if sex == 2
replace female = 0 if sex == 1

tab female 

label var female "Female"

label define fem 1 "Female" 0 "Male"

label values female fem

tab female 

```

```stata



stpm2  i.diabetes i.sex bmi  preop_creat_mg preop_dialysis  smoke  copd  hyperlip  pvd /// 
i.lvfunction  priority prior_pci i.diabetes#i.sex, scale(h) df(5) eform
```

```stata
label define dm 0 "No DM" 1 "DM"

label values diabetes dm 

tab diabetes
```

```stata
// now to run the same model with more meaningful variable labels.


stpm2  i.diabetes i.female age10 bmi5  preop_creat_mg preop_dialysis  smoke  copd  hyperlip  pvd /// 
i.lvfunction  priority prior_pci i.diabetes#i.female i.diabetes#i.lvfunction preop_mi ///
i.diabetes#i.preop_mi , scale(h) df(5) eform

```
